```python
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python

img_grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('greyimage',img_grey)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
(thresh,img_binary) = cv2.threshold(img_grey,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
cv2.imshow('binary_image',img_binary)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imwrite('data/index_gray.jpeg',img_grey)
```
